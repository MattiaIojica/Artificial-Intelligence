```python
import numpy as np
from  sklearn import preprocessing
from sklearn import svm
import matplotlib.image as plt

# In[2]:


def normalize_data(train_data, test_data, type = None):
    if type == 'standard':
        scaler = preprocessing.StandardScaler()
    elif type == 'l1':
        scaler = preprocessing.Normalizer(norm = 'l1')
    elif type == 'l2':
        scaler = preprocessing.Normalizer(norm = 'l2')
    if type is None:
        return train_data, test_data
    else:
        scaler.fit(train_data)
        scaled_train = scaler.transform(train_data)
        scaled_test = scaler.transform(test_data)
        return scaled_train, scaled_test

# In[10]:


def confusion_matrix(y_true, y_pred): 
    num_classes = len(np.unique(y_true))
    conf_matrix = np.zeros((num_classes, num_classes)) 
    
    for i in range(len(y_true)): 
        conf_matrix[int(y_true[i]), int(y_pred[i])] +=1
    return conf_matrix


# In[3]:


def train(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train)
    acc_train = model.score(x_train, y_train)
    acc_test = model.score(x_test, y_test)
    print("accuracy pentru train: {}".format(acc_train))
    print("accuracy pentru test: {}".format(acc_test))

# In[4]:


test_data = []
testfile = open("test.txt", 'r')
lines_test = [x.rstrip("\n") for x in testfile.readlines()]
for i in range(1, len(lines_test)):
    name = lines_test[i].strip('\n')
    img = plt.imread('./test/' + name)
    test_data.append(img)
    
test_data = np.array(test_data)

#print(test_data.shape) (2819, 16, 16, 3)

test_data = test_data.flatten().reshape(2819, 768)

# In[5]:


train_data = []
train_labels = []
trainfile = open("train.txt", 'r')
lines_train = [x.rstrip("\n") for x in trainfile.readlines()]
for i in range(1, len(lines_train)):
    name, label = lines_train[i].split(',')
    img = plt.imread('./train+validation/' + name)
    train_data.append(img)
    train_labels.append(label[0])

train_data = np.array(train_data)

#print (train_data.shape) (8000, 16, 16, 3)

# 16*16*3 = 768, util pentru reshape

train_data = train_data.flatten().reshape(8000, 768)

# In[6]:


val_data = []
val_labels = []
valfile = open("validation.txt", 'r')
lines_val = [x.rstrip("\n") for x in valfile.readlines()]
for i in range(1, len(lines_val)):
    name, label = lines_val[i].split(',')
    img = plt.imread('./train+validation/' + name)
    val_data.append(img)
    val_labels.append(label[0])
    
val_data = np.array(val_data)

#print(val_data.shape) (1173, 16, 16, 3)

val_data = val_data.flatten().reshape(1173, 768)

# In[39]:


scaled_train, scaled_val = normalize_data(train_data, val_data, type = 'l2')
svm_model = svm.SVC(C=17,kernel='rbf')
train(svm_model, scaled_train, scaled_val, train_labels, val_labels)

# In[9]:


scaled_train, scaled_test = normalize_data(train_data, test_data, type = 'l2')

svm_model = svm.SVC(C=17, kernel='rbf')

svm_model.fit(scaled_train, train_labels) 

predicted = svm_model.predict(scaled_test)

testfile = open("test.txt", 'r')
lines_test = [x.rstrip("\n") for x in testfile.readlines()]

sol = open("solutie6.csv", 'w')
count = 0
sol.write("id,label\n")
for i in range(1,len(lines_test)):
    sol.write(f"{lines_test[i]},{predicted[count]}\n")
    count += 1
sol.close()

# In[11]:


scaled_train, scaled_val = normalize_data(train_data, val_data, type = 'l2')
svm_model = svm.SVC(C=17,kernel='rbf')
train(svm_model, scaled_train, scaled_val, train_labels, val_labels)
predicted = svm_model.predict(scaled_val)
print(confusion_matrix(val_labels, predicted))

print(val_labels)
```
