```python
import matplotlib.pyplot as plt

x = np.arange(0, 3 * np.pi, 0.1)

y = np.sin(x)
y1 = np.cos(x)

plt.plot(x, y)
plt.plot(x, y1)
plt.title("sinus")
plt.xlabel("x value")
plt.ylabel("y value")
plt.legend(["sin", "cos"])


plt.show()
```
